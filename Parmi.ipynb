```python
# Program to extract number 
# of rows using Python 
import xlrd 
import numpy as np
# Give the location of the file 
loc = ("data_parmi.xlsx") 
A_OH_I = np.zeros((9584, 3))
A_OH_I_m = np.zeros(9584)
A_OH_U = np.zeros((9584, 3))
A_OH_U_m = np.zeros(9584)
B_OH_I = np.zeros((9584, 3))
B_OH_I_m = np.zeros(9584)
B_OH_U = np.zeros((9584, 3))
B_OH_U_m = np.zeros(9584)
A_P_I = np.zeros((9584, 3))
A_P_I_m = np.zeros(9584)
A_P_U = np.zeros((9584, 3))
A_P_U_m = np.zeros(9584)
B_P_I = np.zeros((9584, 3))
B_P_I_m = np.zeros(9584)
B_P_U = np.zeros((9584, 3))
B_P_U_m = np.zeros(9584)
wb = xlrd.open_workbook(loc)
sheet = wb.sheet_by_index(0) 
sheet.cell_value(0, 0) 
```

```python
for i in range(sheet.nrows-2):
    nuc = int(sheet.cell_value(i+2, 0))
    for j in range(3):
        A_OH_I[nuc-1, j] = 0.0 if sheet.cell_value(i+2, j+1) == 'NA' else sheet.cell_value(i+2, j+1)
        A_OH_U[nuc-1, j] = 0 if sheet.cell_value(i+2, j+4)=='NA' else sheet.cell_value(i+2, j+4)
        B_OH_I[nuc-1, j] = 0 if sheet.cell_value(i+2, j+7)=='NA' else sheet.cell_value(i+2, j+7)
        B_OH_U[nuc-1, j] = 0 if sheet.cell_value(i+2, j+10)=='NA' else sheet.cell_value(i+2, j+10)
        A_P_I[nuc-1, j] = 0 if sheet.cell_value(i+2, j+13)=='NA' else sheet.cell_value(i+2, j+13)
        A_P_U[nuc-1, j] = 0 if sheet.cell_value(i+2, j+16)=='NA' else sheet.cell_value(i+2, j+16)
        B_P_I[nuc-1, j] = 0 if sheet.cell_value(i+2, j+19)=='NA' else sheet.cell_value(i+2, j+19)
        B_P_U[nuc-1, j] = 0 if sheet.cell_value(i+2, j+22)=='NA' else sheet.cell_value(i+2, j+22)
for i in range(9584):
    A_OH_U_m[i] = np.mean(A_OH_U[i, :])
    A_P_U_m[i] = np.mean(A_P_U[i, :])
    A_OH_I_m[i] = np.mean(A_OH_I[i, :])
    A_P_I_m[i] = np.mean(A_P_I[i, :])
    B_OH_U_m[i] = np.mean(B_OH_U[i, :])
    B_P_U_m[i] = np.mean(B_P_U[i, :])
    B_OH_I_m[i] = np.mean(B_OH_I[i, :])
    B_P_I_m[i] = np.mean(B_P_I[i, :]) 
```

```python
import matplotlib.pyplot as plt
x = np.linspace(1, 9584, 9584)
fig, ax= plt.subplots(constrained_layout=True, figsize=(10, 6))
ax.plot(x, A_OH_I_m, 'o', label = 'A_OH')
ax.plot(x, A_P_I_m, 'o', label = 'A_P')
ax.xaxis.set_ticks(np.arange(0, 10000, 500))
ax.legend()
# ax.plot(x, A_OH_U_m, 'o')
# plt.xlim(1, 10000)
```
